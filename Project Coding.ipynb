```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Mental_Health_and_Social_Media_Balance_Dataset.csv")
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Daily Screen Time (hours)'], df['Stress Level (1-10)'], 
            color='orange', marker='x')

plt.title("Screen Time vs Stress Level")
plt.xlabel("Daily Screen Time (hours)")
plt.ylabel("Stress Level (1-10)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Daily Screen Time (hours)'], df['Sleep Quality (1-10)'],
            color='orange', marker='x')

plt.title("Screen Time vs Sleep Quality")
plt.xlabel("Daily Screen Time (hours)")
plt.ylabel("Sleep Quality (1-10)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

How does daily screen timme relate to stress levels??

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Daily_Screen_Time(hrs)'], df['Stress_Level(1-10)'],
            color='orange', marker='x')

plt.title("Screen Time vs Stress Level")
plt.xlabel("Daily Screen Time (hrs)")
plt.ylabel("Stress Level (1-10)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Which platforms perform the highest stress levels?

```python
avg_stress = df.groupby('Social_Media_Platform')['Stress_Level(1-10)'].mean()

plt.figure(figsize=(10, 5))
avg_stress.plot(kind='bar', color='orange')

plt.title("Average Stress Level by Social Media Platform")
plt.xlabel("Platform")
plt.ylabel("Average Stress Level (1-10)")
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Instagram, Youtube, Facebook are the top three sociel media platforms that perform the most stress levels

Does increased screen time impact sleep quality?

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Daily_Screen_Time(hrs)'], df['Sleep_Quality(1-10)'],
            color='orange', marker='x')

plt.title("Screen Time vs Sleep Quality")
plt.xlabel("Daily Screen Time (hrs)")
plt.ylabel("Sleep Quality (1-10)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Yes because screen time before bed negatively impacts sleep quality by suppressing the production of the sleep hormone melatonin due to blue light

4. Can we predict users who are at mental health risk and whats the distribution of daily social media screen time? 

```python
plt.figure(figsize=(12, 6))

plt.hist(df['Daily_Screen_Time(hrs)'], bins=10, color='orange', edgecolor='black')

plt.title("Distribution of Daily Social Media Screen Time", fontsize=16)
plt.xlabel("Daily Screen Time (hours)", fontsize=12)
plt.ylabel("Number of Users", fontsize=12)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

Most people on the daily spend just about 6 hours of screen time and the number of user is 100. In conclusion, yes, screen time can be a predictor or indicator of mental health risk, especially for depression and anxiety in youth, but it's complex: it's less about total hours and more about what you're doing
